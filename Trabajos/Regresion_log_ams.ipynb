```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Datos_limpios_Amsterdam.csv")
df.head(20)
```

```python
df = df[df['review_scores_cleanliness'] != 'Nothing']
df = df[df['host_is_superhost'] != 'Without information']
df = df[df['host_acceptance_rate'] != 'Without information']
```

```python
df['host_acceptance_rate'] = df['host_acceptance_rate'].str.rstrip('%').astype(float) / 100
df['host_response_rate'] = df['host_response_rate'].str.rstrip('%').astype(float) / 100
```

```python
Vars_Indep= df[['host_acceptance_rate','host_response_rate','review_scores_rating']]
Var_Dep= df['host_is_superhost']
```

```python
X= Vars_Indep
y= Var_Dep
```

```python
df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=None)
```

```python
escalar= StandardScaler()
```

```python
X_train= escalar.fit_transform(X_train)
X_test= escalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
algoritmo= LogisticRegression()
```

```python
algoritmo.fit(X_train, y_train)
```

```python
y_pred= algoritmo.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
matriz= confusion_matrix(y_test, y_pred)
print('Matriz de ConfusiÃ³n:')
print(matriz)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test,y_pred,average="binary", pos_label="t")
print("PrecisiÃ³n del modelo:")
print(precision)
```

```python
from sklearn.metrics import accuracy_score
exactitud= accuracy_score(y_test, y_pred)
print('Exactitud del modelo:')
print(exactitud)
```

```python
from sklearn.metrics import recall_score
sensibilidad= recall_score(y_test, y_pred, average="binary",pos_label="f")
print('Sensibilidad del modelo:')
print(sensibilidad)
```
